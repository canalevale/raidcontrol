```python
import torch
import torch.nn as nn
import sys
import os
# Agrega el directorio padre (Source) al sys.path
sys.path.append(os.path.abspath('..'))
from raidcontrol.NumberCNN_OCRClass import DigitCNN
```

```python
def main():
    ckpt_path = "digit_cnn_64_v4.pt"
    onnx_path = "digit_cnn_64_v4.onnx"

    ckpt = torch.load(ckpt_path, map_location="cpu")
    model = DigitCNN()

    if isinstance(ckpt, dict) and "model_state" in ckpt:
        model.load_state_dict(ckpt["model_state"])
    else:
        model.load_state_dict(ckpt)

    model.eval()

    dummy = torch.randn(1, 1, 56, 56)

    torch.onnx.export(
        model,
        dummy,
        onnx_path,
        input_names=["input"],
        output_names=["logits"],
        opset_version=12,
        dynamic_axes={
            "input": {0: "batch"},
            "logits": {0: "batch"},
        },
        dynamo=False,
    )

    print("Saved ONNX model to:", onnx_path)
```

```python
main()
```
